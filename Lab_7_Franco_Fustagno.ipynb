```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_1samp
from scipy import stats
```

```python
#Example for Beerios
#H0= 9 grams, H1 != 9
beerios = [9.16, 8.57, 9.54, 8.70, 9.33, 9.43, 8.39, 9.47, 8.66, 9.01]
```

```python
#Normality testing
%pip install pingouin
import pingouin as pg
```

```python
#normality check
pg.qqplot(beerios)
plt.show()
```

```python
ttest_1samp(a=beerios,popmean=9)
```

```python
t_stat, p_value= ttest_1samp(a=beerios, popmean=9)
print(p_value)
```

```python
if p_value < .05:
  print('Reject null hypothesis')
else:
  print('Fail to reject null hypothesis')
```

```python
ttest_1samp(beerios, popmean=9).confidence_interval(confidence_level=0.95)
```

```python
#Example Golf club
#Hypothesis  H0= 0.82, H1 > 0.82
Club= [0.8411, 0.8191, 0.8182, 0.8125, 0.8750, 0.8580, 0.8532, 0.8483, 0.8276, 0.7983, 0.8042, 0.8730, 0.8282, 0.8359, 0.8660]
```

```python
#Normal approximation of Poisson distribution
import scipy.stats as stats

pop= stats.poisson.rvs(loc=18, mu= 35, size= 1500)
samp_1= stats.poisson.rvs(loc=18, mu= 35, size= 60) 
samp_2= stats.poisson.rvs(loc=18, mu= 30, size= 60)
```

```python
pop.mean()
```

```python
samp_1.mean()
```

```python
samp_2.mean()
```

```python
ttest_1samp(samp_1, popmean=pop.mean()).confidence_interval(confidence_level=0.95)
```

```python
ttest_1samp(samp_2, popmean=pop.mean()).confidence_interval(confidence_level=0.95)
```

```python
pg.qqplot(samp_1)
plt.show()
```

```python
ttest_1samp(a=samp_1,popmean=pop.mean())
```

```python
ttest_1samp(a=samp_2,popmean=pop.mean())
```

**Exponential Distribution**

```python
#sample size is reduced to fit the normality test.
```

```python
pop_exp= stats.expon.rvs(loc=18, scale= 2, size= 1500)
samp_1_exp= stats.expon.rvs(loc=18, scale= 2, size= 10) 
samp_2_exp= stats.expon.rvs(loc=18, scale= 3, size= 10)
```

```python
#normality test
pg.qqplot(samp_1_exp)
plt.show()
```

```python
pg.qqplot(samp_2_exp)
plt.show()
```

```python
pop_exp.mean()
```

```python
samp_1_exp.mean()
```

```python
samp_2_exp.mean()
```

```python
t_stat, p_value= ttest_1samp(a=samp_1_exp, popmean= pop_exp.mean())
if p_value < 0.05:
  print("We reject the null hypothesis")
else:
  print("We fail to rejet the null hypothesis")
```

```python
t_stat, p_value= ttest_1samp(a=samp_2_exp, popmean= pop_exp.mean())
if p_value < 0.05:
  print("We reject the null hypothesis")
else:
  print("We fail to rejet the null hypothesis")
```
